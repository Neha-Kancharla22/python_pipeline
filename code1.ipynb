```python
#first question
import pandas as pd
import random
def generate_phone():
    return ''.join([str(random.randint(0, 9)) for _ in range(10)])
df1=pd.read_csv(r"C:\Users\Neha\Downloads\us_customer_data 1.csv")
df1.info()
print(df1.duplicated().sum())
df1.drop_duplicates
df1['phone'] = '+91-'+df1['phone'].apply(lambda x: generate_phone() if pd.isna(x) or x == '' else x)
df1['phone']=df1['phone'].str.replace(r'\D', '',regex=True)
df1['phone'] = '+91-'+ df1['phone']
df1['email']=df1['email'].fillna('nehaa@example.com')
print(df1)


```
